```python
# Import the necessary libraries 
import pandas as pd
```

```python
# load one csv for the coil 10059b4 at the stage B4
df_B4 = pd.read_csv('SignalExport/10060b4.csv')
df_B4
```

```python
# load one csv for the coil 10059b4 at the stage B5
df_B5 = pd.read_csv('SignalExport/10060b5.csv')
df_B5
```

```python
# B4 turn it into the list and split it by semicolon
df_B4=list(df_B4.columns)[0].split(";")
#df_B4
```

```python
# B5 turn it into the list and split it by semicolon
df_B5=list(df_B5.columns)[0].split(";")
#df_B5
```

```python
# B4 Finding index for the length and width
li_B4=df_B4.index("Lengthpoints:") 
wi_B4=df_B4.index("Values")
```

```python
# B5 Finding index for the length and width
li_B5=df_B5.index("Lengthpoints:") 
wi_B5=df_B5.index("Values")
```

```python
# B4 Slicing the dataframe into 4 lists 
coil_number_B4=df_B4[0]
coil_stage_B4 = df_B4[1]
length_B4=df_B4[li_B4+1:wi_B4]   
width_B4=df_B4[wi_B4+1:-1]
```

```python
# B5 Slicing the dataframe into 4 lists 
coil_number_B5=df_B5[0]
coil_stage_B5 = df_B5[1]
length_B5=df_B5[li_B5+1:wi_B5]   
width_B5=df_B5[wi_B5+1:-1]
```

```python
# B4 Checking the number of observations in the lists
print(len(coil_number_B4))
print(len(length_B4))
print(len(width_B4))
```

```python
# B5 Checking the number of observations in the lists
print(len(coil_number_B5))
print(len(length_B5))
print(len(width_B5))
```

```python
# B4 Turning length and width measurements from a string into a float 
length_B4 = [float(i) for i in length_B4]
width_B4 = [float(i) for i in width_B4]
```

```python
# B5 Turning length and width measurements from a string into a float 
length_B5 = [float(i) for i in length_B5]
width_B5 = [float(i) for i in width_B5]
```

```python
# B4 Duplicate the coil number for the number of observations in the length 
mult_coil_values_B4= [coil_number_B4 for i in range(len(length_B4))]
```

```python
# B5 Duplicate the coil number for the number of observations in the length 
mult_coil_values_B5= [coil_number_B5 for i in range(len(length_B5))]
```

```python
# B4 Duplicate the coil stage id(B4 or B5) for the number of observations in the length 
mult_coil_stage_values_B4 = [coil_stage_B4 for i in range(len(length_B4))]
```

```python
# B5 Duplicate the coil stage id(B4 or B5) for the number of observations in the length 
mult_coil_stage_values_B5 = [coil_stage_B5 for i in range(len(length_B5))]
```

```python
# B4 Create an empty dataframe 
my_B4_df = pd.DataFrame()
```

```python
# B5 Create an empty dataframe 
my_B5_df = pd.DataFrame()
```

```python
# B4 Turn in the lists into the series  
mult_coil_values_B4 = pd.Series(mult_coil_values_B4)
coil_stage_values_B4 = pd.Series(mult_coil_stage_values_B4)
length_values_B4 = pd.Series(length_B4)
width_values_B4 = pd.Series(width_B4)
```

```python
# B5 Turn in the lists into the series  
mult_coil_values_B5 = pd.Series(mult_coil_values_B5)
coil_stage_values_B5 = pd.Series(mult_coil_stage_values_B5)
length_values_B5 = pd.Series(length_B5)
width_values_B5 = pd.Series(width_B5)
```

```python
# B4 Create columns and fill them with the series 
my_B4_df['coil_number'] = mult_coil_values_B4
my_B4_df['coil_stage'] = mult_coil_stage_values_B4
my_B4_df['length'] = length_values_B4
my_B4_df['width'] = width_values_B4
```

```python
# B5 Create columns and fill them with the series 
my_B5_df['coil_number'] = mult_coil_values_B5
my_B5_df['coil_stage'] = mult_coil_stage_values_B5
my_B5_df['length'] = length_values_B5
my_B5_df['width'] = width_values_B5
```

```python
#B4 Check the dataframe
my_B4_df.head(10)
```

```python
#B5 Check the dataframe
my_B5_df.head(10)
```

```python
# B4 Drop the zeros
my_B4_df = my_B4_df[(my_B4_df != 0).all(1)]
my_B4_df
```

```python
# B5 Drop the zeros
my_B5_df = my_B5_df[(my_B5_df != 0).all(1)]
my_B5_df.head(15)
```

```python
new_df_B4 = my_B4_df.loc[(my_B4_df['length'] >= 140) & (my_B4_df['length'] <= 170)]
new_df_B4.shape
```

```python
new_df_B5 = my_B5_df.loc[(my_B5_df['length'] >= 140) & (my_B5_df['length'] <= 170)]
new_df_B5.shape
```

```python
#Plot the graph for B4 coil
import matplotlib.pyplot as plt
plt.plot(my_B4_df.length,my_B4_df.width,"g")
plt.plot(my_B5_df.length,my_B5_df.width,"r")
plt.rcParams["figure.figsize"] = (8,10)
plt.show()
```

```python
#Plot the graph for B4 coil
import matplotlib.pyplot as plt
plt.plot(new_df_B4.length,new_df_B4.width,"g",label ="B4")
plt.plot(new_df_B5.length,new_df_B5.width,"r",label = "B5")
plt.rcParams["figure.figsize"] = (8,10)
plt.legend()
plt.show()
```
